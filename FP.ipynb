### Data Tingkat Pengangguran di Indonesia periode 2006 - 2022

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('Data Training-1.csv')
```

```python
#dataset
display(df)
```

```python
#info data
print("\nInfo Dataset")
print(df.info())
```

### EDA

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Periode', y='Total', hue='Bulan', marker='o')
plt.title('Tren Jumlah Total per Tahun (Februari vs Agustus)')
plt.ylabel('Jumlah Orang Pengangguran')
plt.show()
```

```python
cols_pendidikan = ['Tidak/belum pernah sekolah', 'Tidak/belum tamat SD', 'SD', 
                   'SLTP', 'SLTA Umum/SMU', 'SLTA Kejuruan/SMK', 
                   'Akademi/Diploma', 'Universitas']
```

```python
mean_values = df[cols_pendidikan].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=mean_values.values, y=mean_values.index, hue=mean_values.index, palette='viridis',legend=False)
plt.title('Rata-rata Jumlah per Tingkat Pendidikan')
plt.xlabel('Rata-rata Jumlah')
plt.ylabel('Pendidikan')
plt.show()
```
